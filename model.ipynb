```python
# Cell 2 - Imports
from ultralytics import YOLO
import torch
import os

print("CUDA Available:", torch.cuda.is_available())
```

```python
# Cell 3 - Dataset path check
DATA_YAML = "yolo_dataset/data.yaml"
print("Dataset Found:", os.path.exists(DATA_YAML))
```

```python
# Cell 4 - Load pretrained YOLO model
# balanced speed + accuracy
model = YOLO("yolov8s.pt")
```

```python
# Cell 5 - Train model professionally with augmentations
results = model.train(
    data=DATA_YAML,
    epochs=80,
    imgsz=640,
    batch=16,
    patience=20,
    workers=2,
    device=0,

    # optimizer
    optimizer="AdamW",
    lr0=0.001,
    weight_decay=0.0005,

    # augmentations
    hsv_h=0.015,
    hsv_s=0.7,
    hsv_v=0.4,

    degrees=20,         # rotation
    translate=0.15,     # shift
    scale=0.50,         # near/far simulation
    shear=8,
    perspective=0.0015,

    fliplr=0.5,
    flipud=0.0,

    mosaic=1.0,
    mixup=0.15,
    copy_paste=0.15,

    erasing=0.25,       # partial hide
    crop_fraction=1.0,

    close_mosaic=10,
    plots=True,
    verbose=True
)
```

```python
# Cell 6 - Validate trained model
metrics = model.val()

print("mAP50-95 :", metrics.box.map)
print("mAP50    :", metrics.box.map50)
print("Precision:", metrics.box.mp)
print("Recall   :", metrics.box.mr)
```

```python
# Cell 7 - Predict on validation images only (NO webcam)
model.predict(
    source="yolo_dataset/images/val",
    conf=0.35,
    imgsz=640,
    save=True,
    show=True
)
```
